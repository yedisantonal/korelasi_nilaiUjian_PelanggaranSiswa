```python
def mewarnai_max(col):
    warna = "background-color:red;"
    warni = "background-color:green;"
    nol =""
    
    nilai_max = col.max()
    nilai_min = col.min()

    return [warna if v == nilai_max else ( warni if v == nilai_min else nol)  for v in col]
    
df.style.apply(mewarnai_max, subset=['PAI', 'PKN'], axis=0)
```

# Deteksi Oulier

```python
df_mean = df['PAI'].mean()
df_std = df['PAI'].std()
limit_atas = df_mean - (3*df_std)
limit_bawah = df_mean + (3*df_std)
limit = 3*df_std
limit
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.hist(df.PAI, bins=100)

ax.axvline(limit_atas, color ='r')
ax.axvline(limit_bawah, color ='r')
plt.show()
```

```python
df_mean_pkn = df['PKN'].mean()
df_std_pkn = df['PKN'].std()
limit_atas_pkn = df_mean_pkn - (3*df_std_pkn)
limit_bawah_pkn = df_mean_pkn + (3*df_std_pkn)
limit_pkn = 3*df_std_pkn
limit_pkn
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.hist(df.PKN, bins=100)

ax.axvline(limit_atas_pkn, color ='r')
ax.axvline(limit_bawah_pkn, color ='r')
plt.show()
```

```python
df_mean_pl = df['Pelanggaran'].mean()
df_std_pl = df['Pelanggaran'].std()
limit_atas_pl = df_mean_pl - (3*df_std_pl)
limit_bawah_pl = df_mean_pl + (3*df_std_pl)
limit_pl = 3*df_std_pl
limit_pl
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.hist(df.Pelanggaran, bins=100)

ax.axvline(limit_atas_pl, color ='r')
ax.axvline(limit_bawah_pl, color ='r')
plt.show()
```
